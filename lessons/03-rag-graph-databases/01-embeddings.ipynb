```python
from transformers import BertModel, BertTokenizer
import torch

def get_3d_word_embedding(word):
    # Load pre-trained BERT model and tokenizer
    model_name = 'bert-base-uncased'
    tokenizer = BertTokenizer.from_pretrained(model_name)
    model = BertModel.from_pretrained(model_name)

    # Tokenize the input word
    inputs = tokenizer(word, return_tensors='pt')

    # Get the embeddings from the model
    with torch.no_grad():
        outputs = model(**inputs)

    # The embeddings are in the last hidden state
    embeddings = outputs.last_hidden_state

    # Extract the embedding for the input word (first token)
    word_embedding = embeddings[0][0].numpy()

    # Select the first three dimensions for a 3D representation
    word_embedding_3d = word_embedding[:3]

    return word_embedding_3d

# Example usage
word = input("Enter a word: ")
embedding_3d = get_3d_word_embedding(word)

print("\n")
print(f"3D Embedding for the word '{word}':")
print(embedding_3d)

```

```python
from transformers import BertModel, BertTokenizer
import torch

def get_word_embedding(word):
    # Load pre-trained BERT model and tokenizer
    model_name = 'bert-base-uncased'
    tokenizer = BertTokenizer.from_pretrained(model_name)
    model = BertModel.from_pretrained(model_name)

    # Tokenize the input word
    inputs = tokenizer(word, return_tensors='pt')

    # Get the embeddings from the model
    with torch.no_grad():
        outputs = model(**inputs)

    # The embeddings are in the last hidden state
    embeddings = outputs.last_hidden_state

    # Return the embedding of the input word (first token)
    word_embedding = embeddings[0][0].numpy()

    return word_embedding

# Example usage
word = input("Enter a word: ")
embedding = get_word_embedding(word)

print(f"Embedding for the word '{word}':")
print(embedding)
```
